```python
''' This won't let u Modify or add things in the text 
    but it will split the single word the way you want'''

from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("Lemme",
    [{ORTH : "Lem"},
     {ORTH : "me"}
     ])

doc = nlp("Lemme go and grab some ice-cream")
tokens = [token.text for token in doc]
tokens
```
